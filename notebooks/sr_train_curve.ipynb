```python
import matplotlib.pyplot as plt
```

```python
# hand-recorded data
epoch = [3.5, 4.5, 10]
sr_text = [0.28, 0.31, 0.33]
sr_visual = [0.25, 0.29, 0.31]
```

```python
font = {
    "fontname": "Times New Roman",
    "size": 10,
    "math_fontfamily": "stix",
}


# Create the figure and the axes
fig, ax = plt.subplots(figsize=(5.09, 2), dpi=300, layout="constrained")

# Plot data
ax.plot(
    epoch,
    sr_text,
    color="darkred",
    marker="o",
    linestyle="-",
    linewidth=2,
    label="SR (Textual Conditioning)",
)
ax.plot(
    epoch,
    sr_visual,
    color="red",
    marker="o",
    linestyle="-",
    linewidth=2,
    label="SR (Visual Conditioning)",
)

# Set the plot title and labels
# ax.set_title('SR Conditioning over Epochs', fontsize=15, color='darkred')
ax.set_xlabel("Epoch", **font)
ax.set_ylabel("SR", **font)

ax.tick_params(labelsize=10)
for label in ax.xaxis.get_ticklabels():
    label.set_family("Times New Roman")
    
for label in ax.yaxis.get_ticklabels():
    label.set_family("Times New Roman")

# Set the tick colors
# ax.tick_params(axis='x', colors='darkred')
# ax.tick_params(axis='y', colors='darkred')

# Add a legend
ax.legend(prop={"family": "Times New Roman", "size": 10})

# Show the plot
plt.savefig(
    "../outputs/calvin/images/sr_train_curve.pdf", bbox_inches="tight", pad_inches=0.01
)
plt.show()
```
