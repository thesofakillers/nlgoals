```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
baseline_sim_matrix = np.load(
    "../outputs/calvin/cclipt_eval/cclipt-s42/similarity-matrix-256.npy"
)
ablated_sim_matrix = np.load(
    "../outputs/calvin/cclipt_eval/cclipt-s42-ablated/similarity-matrix-256.npy"
)
```

```python
cmap='Reds'
```

```python
f, (ax1, ax2) = plt.subplots(
    1, 2, figsize=(5.09, 4), sharex=True, sharey=True, layout="compressed", dpi=600
)
font = {
    "fontname": "Times New Roman",
    "size": 10,
    "math_fontfamily": "stix",
}

im1 = ax1.imshow(baseline_sim_matrix, cmap=cmap, vmin=-1, vmax=1)
ax1.set_title("Baseline", **font)
ax1.set_ylabel(r"$\mathbf{v}_{tg}$", **font)
ax1.set_yticks([])
ax1.set_xticks([])
ax1.set_xlabel(r"$\mathbf{v}_{vg}$", **font)

im2 = ax2.imshow(ablated_sim_matrix, cmap=cmap, vmin=-1, vmax=1)
ax2.set_title("Ablation", **font)
ax2.set_xlabel(r"$\mathbf{v}_{vg}$", **font)

cbar = f.colorbar(im1, ax=[ax1, ax2], orientation="horizontal", pad=-0.1)
cbar.set_label("Similarity", **font)
cbar.ax.xaxis.set_label_position('top')
cbar.ax.tick_params(labelsize=8)
for label in cbar.ax.xaxis.get_ticklabels():
    label.set_family("serif")

plt.savefig(
    "../outputs/calvin/images/sim_matrices.pdf",
    bbox_inches="tight",
    pad_inches=0,
)
plt.show()
```
